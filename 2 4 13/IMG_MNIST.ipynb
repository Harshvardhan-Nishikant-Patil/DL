```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mat
import keras 
import tensorflow as tf
from tensorflow.keras.layers import Dense,Dropout,BatchNormalization,MaxPooling2D,Conv2D,Flatten
from tensorflow.keras.models import Sequential
```

```python
dataset=pd.read_csv('../LP-IV-datasets/MNIST(Ass2&3)/mnist_train.csv')
dataset
```

```python
x_train=dataset.drop(['label'],axis=1).values
x_train=x_train.reshape((60000,28,28,1))
x_train=x_train.astype('float32')/255
x_train
```

```python
y_train=dataset['label']
y_train
```

```python
test=pd.read_csv('../LP-IV-datasets/MNIST(Ass2&3)/mnist_test.csv')
test
```

```python
x_test = test.drop(['label'], axis=1).values
x_test = x_test.reshape((10000, 28, 28, 1))
x_test = x_test.astype('float32') / 255
x_test
```

```python
y_test=test['label']
y_test
```

```python
model=Sequential([
    Conv2D(32,(3,3),input_shape=(28,28,1),activation='relu'),
    BatchNormalization(),
    Conv2D(32,(3,3),activation='relu'),
    MaxPooling2D((2,2)),
    Flatten(),
    Dense(128,activation='relu'),
    Dropout(0.2),
    Dense(10,activation='softmax')
])
model.compile(loss='sparse_categorical_crossentropy',metrics=['accuracy'],optimizer='adam')
model.summary()
```

```python
model.fit(x_train,y_train,epochs=11,batch_size=64)
```

```python
los,acc=model.evaluate(x_test,y_test)
print(acc)
```
